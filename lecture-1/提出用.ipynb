```python
import time
import requests
from bs4 import BeautifulSoup
import json
```

```python
#urlの指定
url = 'https://github.com/orgs/google/repositories'

#取得したリポジトリー名を格納するためのリストを作る
repository_name =[]
repository_lang = []
repository_star = []

#ページ数が94ページあることがわかるので以下のプログラムを94ページ分巡回するようにする
#githubはページ番号がurlの末尾についており、2ページ目は?page=2が追加されている（1ページ目には?page=1はついていない）
#nページ目は?page=nとなる、pythonのrange関数は終点を含まないので終点+1をする

try:
    time.sleep(1)
    #request.get()を使い、urlのhtmlを取得する
    responce = requests.get(url)
    #responceのcontent(中身)であるhtmlをBeautifuiSoupを使って分析する、'html.parser'を入れることでそのhtmlを分析できる
    soup = BeautifulSoup(responce.content,'html.parser')
    #分析したhtml内のリポジトリーの名前が含まれる部分を選択して取得する
    #soup.findとsoup.selectの2種類の方法があり、リポジトリ名と使用言語のデータは、すべて一つの大きなJSONデータの塊として、HTML内の一つの<script>タグの中に格納されているため
    # 一つの要素（今回はjsonデータ）を取得するsoup.findの方が良い
    script_tag = soup.find('script', {'data-target': 'react-app.embeddedData', 'type': 'application/json'})
    data = json.loads(script_tag.string)
    total_pages = data.get("payload", {}).get("orgReposPageRoute", {}).get("pageCount")
except requests.exceptions.HTTPError as e:
        print('HTTPエラーが発生しました:', e)
except requests.exceptions.RequestException as e:
        print("リクエストエラーが発生しました:", e)

for page in range (1, total_pages+1):
    if page == 1:
        url == url
    else:
        url == f"{url}?page={page}"

    try:
        time.sleep(1)
        #request.get()を使い、urlのhtmlを取得する
        responce = requests.get(url)
        #responceのcontent(中身)であるhtmlをBeautifuiSoupを使って分析する、'html.parser'を入れることでそのhtmlを分析できる
        soup = BeautifulSoup(responce.content,'html.parser')
        #分析したhtml内のリポジトリーの名前が含まれる部分を選択して取得する
        #soup.findとsoup.selectの2種類の方法があり、リポジトリ名と使用言語のデータは、すべて一つの大きなJSONデータの塊として、HTML内の一つの<script>タグの中に格納されているため
        # 一つの要素（今回はjsonデータ）を取得するsoup.findの方が良い
        script_tag = soup.find('script', {'data-target': 'react-app.embeddedData', 'type': 'application/json'})
        

        if script_tag:
            # 2. scriptタグの中身 (JSON文字列) を抽出
            json_string = script_tag.string 
            
            # 3. JSON文字列をPythonの辞書にパース
            data = json.loads(json_string)
            
            # 4. 辞書からリポジトリのリストを抽出
            # キーの階層: payload -> orgReposPageRoute -> repositories
            repos_list = data.get("payload", {}).get("orgReposPageRoute", {}).get("repositories", [])
            
            # 5. リストから個々のリポジトリ名と主要言語を抽出
            for repo in repos_list:
                repository_name.append(repo.get("name"))
                
                # リポジトリの主要言語も抽出 (例: Python, C++)
                primary_lang = repo.get("primaryLanguage", {}).get("name")
                if primary_lang:
                    repository_lang.append(primary_lang)
                
                starsCount = repo.get("starsCount")
                if starsCount:
                    repository_star.append(starsCount)
            

        # 6. ページ全体の主要言語（HTMLのlang属性）を取得
        if soup.html and 'lang' in soup.html.attrs:
            page_language = soup.html['lang']


    

        responce.raise_for_status()  #HTTPエラーが発生した場合に例外を発生させる
    except requests.exceptions.HTTPError as e:
        print('HTTPエラーが発生しました:', e)
    except requests.exceptions.RequestException as e:
        print("リクエストエラーが発生しました:", e)
else:
    print("webページに正常にアクセスできました。")
    
print("ウェブスクレイピングに処理が完了しました")
print(f"ページ数は{total_pages}ページです。")
print(f"リポジトリーの数は{len(repository_name)}個です。")
print("リポジトリー名:",repository_name)
print("主要言語:",repository_lang)
print("スターの数:",repository_star)
```

```python
path = ''
db_name = '提出用.db'
try:
    #DB接続オブジェクトの作成
    conn = sqlite3.connect(path + db_name)

    #SQL(RDBを操作するための言語)カーソルオブジェクトを取得
    cur = conn.cursor()

    #SQL文の作成
    #テーブルの作成
    sql = 'create table repository (リポジトリ名 INT, 使用言語 INT, スター数 REALL);'

    #sql文の実行
    cur.execute(sql)

except sqlite3.Error as e:
    print("エラーが発生しました",e)

finally:
    #DBへの接続を閉じる
    conn.close()
```

```python
path = ''
db_name = '提出用.db'

try:
    #DB接続オブジェクトの作成
    conn = sqlite3.connect(path + db_name)

    #SQL(RDBを操作するための言語)カーソルオブジェクトを取得
    cur = conn.cursor()

    #SQL文の作成
    #データの挿入
    #リストに入っているリポジトリ名, 使用言語, スター数はそのままでは挿入できないのでプレ-スホルダーを利用する
    #()内の?がプレースホルダーの役割をしている
    sql = "INSERT INTO repository (リポジトリ名, 使用言語, スター数) VALUES (?, ?, ?)"
    #挿入をリポジトリー名の数だけ行う
    for i in range(len(repository_name)):
        #リストの中のi番目のデータをプレスホルダーに挿入する
        data_to_insert = (repository_name[i], repository_lang[i], repository_star[i]);

        #sql文の実行
        #プレスホルダーの内容をデータベースに挿入する
        cur.execute(sql, data_to_insert)

    #変更をDBに反映させる
    conn.commit()
    print(f"{len(repository_name)}個のデータを挿入しました。")

except sqlite3.Error as e:
    print('エラーが発生しました',e)

finally:
    #DBへの接続を閉じる
    conn.close()
```
