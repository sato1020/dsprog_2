```python
import time
import requests
from bs4 import BeautifulSoup
import json
```

```python
#urlの指定
url = 'https://github.com/orgs/google/repositories'

#取得したリポジトリー名を格納するためのリストを作る
repository_name =[]
repository_lang = []

try:
    time.sleep(1)
    #request.get()を使い、urlのhtmlを取得する
    responce = requests.get(url)
    #responceのcontent(中身)であるhtmlをBeautifuiSoupを使って分析する、'html.parser'を入れることでそのhtmlを分析できる
    soup = BeautifulSoup(responce.content,'html.parser')
    #分析したhtml内のリポジトリーの名前が含まれる部分を選択して取得する
    #soup.findとsoup.selectの2種類の方法があり、リポジトリ名と使用言語のデータは、すべて一つの大きなJSONデータの塊として、HTML内の一つの<script>タグの中に格納されているため
    # 一つの要素（今回はjsonデータ）を取得するsoup.findの方が良い
    script_tag = soup.find('script', {'data-target': 'react-app.embeddedData', 'type': 'application/json'})
    

    if script_tag:
        # 2. scriptタグの中身 (JSON文字列) を抽出
        json_string = script_tag.string 
        
        # 3. JSON文字列をPythonの辞書にパース
        data = json.loads(json_string)
        
        # 4. 辞書からリポジトリのリストを抽出
        # キーの階層: payload -> orgReposPageRoute -> repositories
        repos_list = data.get("payload", {}).get("orgReposPageRoute", {}).get("repositories", [])
        
        # 5. リストから個々のリポジトリ名と主要言語を抽出
        for repo in repos_list:
            repository_name.append(repo.get("name"))
            
            # リポジトリの主要言語も抽出 (例: Python, C++)
            primary_lang = repo.get("primaryLanguage", {}).get("name")
            if primary_lang:
                repository_lang.append(primary_lang)

    # 6. ページ全体の主要言語（HTMLのlang属性）を取得
    if soup.html and 'lang' in soup.html.attrs:
        page_language = soup.html['lang']

    #主要なプログラミング言語を取得
    

    responce.raise_for_status()  #HTTPエラーが発生した場合に例外を発生させる
except requests.exceptions.HTTPError as e:
    print('HTTPエラーが発生しました:', e)
except requests.exceptions.RequestException as e:
    print("リクエストエラーが発生しました:", e)
else:
    print("webページに正常にアクセスできました。")
finally:
    print("ウェブスクレイピングに処理が完了しました")
    print("リポジトリー名:",repository_name)
    print("主要言語:",repository_lang)
```
